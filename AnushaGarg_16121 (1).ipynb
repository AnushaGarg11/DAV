<h3>(b)

```python
def remove_outliers_iqr(dataframe):
    Q1 = dataframe.quantile(0.25)
    Q3 = dataframe.quantile(0.75)
    IQR = Q3 - Q1
    return dataframe[~((dataframe < (Q1 - 1.5 * IQR)) | (dataframe > (Q3 + 1.5 * IQR))).any(axis=1)]
```

<h1>QUESTION 6<h1>
<h3>(family.csv)

```python
import pandas as pd
import numpy as np
data_csv = pd.read_csv("family.csv")
data = pd.DataFrame(data_csv)
print(f"Data of family.csv file : \n{data}")
salaries = data.groupby("FamilyName").groups
```

<h3>a)<h3>

```python
# Family Gross Income
family_data = {}
for i in salaries.keys():
    sal_data = data.loc[salaries[i]]
    family_data[i] = sal_data["MonthlyIncome(Rs.)"].sum()
print(f"Family Gross Income : \n{family_data}")
```

<h3>b)<h3>

```python
# Max and Min Salary of all Families
for i in salaries:
    sal_max = data.loc[salaries[i]].max()
    sal_min = data.loc[salaries[i]].min()
    print(f"*****Maximum Salary of family : \n{sal_max}")
    print(f"*****Minimum Salary of family : \n{sal_min}")
    print()
```

<h3>c)<h3>

```python
# Those Family their all members income less then 80000
print("Family their all members income less then 80000")
for i in salaries:
    slr = data.loc[salaries[i]]
    slr_data = slr.where(slr["MonthlyIncome(Rs.)"]<80000).sum(numeric_only=True)
    print(slr_data)
```

<h3>d)<h3>

```python
# Average Income of Shah Family
average = 0
value = data.loc[salaries["Shah"]]
con = value["Gender"] == "Female"
for i in con.keys():
    if(con[i]):
        sal = value.loc[i]
        average += sal["MonthlyIncome(Rs.)"]
average = average/len(value)
print(f"Average Income of Shah Family : {average}")
```

<h3>e)<h3>

```python
# Total income of those members their income greater than 60000
for i in salaries:
    slr = data.loc[salaries[i]]
    if (slr["MonthlyIncome(Rs.)"]>60000).all():
        slr_data = slr.where((slr["MonthlyIncome(Rs.)"]>60000))
        print(f"Total Income : \n{slr_data.sum(numeric_only=True)}")
```

<h3>f)<h3>

```python
# Average Monthly Income of All Females
gender = data.groupby("Gender").groups
average = data.loc[gender['Female']].sum(numeric_only=True)/len(gender["Female"])
print(f"Average of Female Income : {average}")
```

<h3>g)<h3>

```python
# Delete wich rows their income is less than average income
average = data["MonthlyIncome(Rs.)"].mean()
sala_data = data.where((data["MonthlyIncome(Rs.)"]>average)).dropna()
print(f"This is which family member their income greater then average income : \n{sala_data}")
```

```python
import pandas as pd
import numpy as np
data_csv = pd.read_csv("earthquakes.csv")
data = pd.DataFrame(data_csv)
print(data)
```

<h3>a)<h3>

```python
# All the earthquakes in Japan with a magType of mb and a magnitude of 4.9 or greater.
country = data.groupby("parsed_place").groups
cou_data = data.loc[country["Japan"]]
magtype_data = cou_data.groupby('magType').groups
m_data = cou_data.loc[magtype_data['mb']]
mag = m_data.where(m_data['mag']>=4.9).dropna()
print(f"magType 'mb' and magnitude of 4.9 or greater : \n{mag}")
```

<h3>(b)<h3>

```python
mag_data = data.groupby("magType").groups
bin_data = pd.cut(mag_data['ml'],bins=[0,1,2,3,4,5,6])
print(f"Number of bin : {len(bin_data)}")
```

<h3>(c)

```python
crosstab_result = pd.crosstab(data['tsunami'], data['magType'],
values=data['mag'], aggfunc='max', margins=True, margins_name='Max_Magnitude')
print("\nc. Crosstab with the maximum magnitude for each combination of tsunami and magType:")
print(crosstab_result)
```

<h3>(iris.data)</h3>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.stats import mode
from scipy import stats
```

<h3>(a)

```python
iris = load_iris()
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_data['target'] = iris.target
iris_data.info()
```

<h3>(b)

```python
missing_values = iris_data.isnull().sum()
missing_values
```

<h3>(c)

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=iris_data)
plt.title('Class Frequency in Iris Dataset')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```

<h3>(d)

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='sepal length (cm)', y='petal length (cm)', data=iris_data)
plt.title('Scatter Plot: Petal Length vs Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

<h3>(e)

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(iris_data['petal width (cm)'], shade=True)
plt.title('Density Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.show()
```

<h3>(f)

```python
sns.pairplot(iris_data, hue='target', diag_kind='kde')
plt.suptitle('Pairwise Bivariate Distribution in Iris Dataset', y=1.02)
plt.show()
```

<h3>(g)

```python
plt.figure(figsize=(8, 6))
sns.heatmap(iris_data[['sepal length (cm)', 'sepal width (cm)']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap: Sepal Length vs Sepal Width')
plt.show()
```

<h3>(h)

```python
numeric_features = iris_data.drop('target', axis=1)
numeric_stats = pd.DataFrame({
    'Mean': numeric_features.mean(),
    'Mode': mode(numeric_features).mode[0],
    'Median': numeric_features.median(),
    'Std Dev': numeric_features.std(),
    'Confidence Interval (95%)': [stats.norm.interval(0.95, loc=mean, scale=stats.sem(numeric_features[col]))
                                   for col, mean in numeric_features.mean().iteritems()],
    'Standard Error': numeric_features.sem()
})
print("\nNumeric Feature Statistics:")
print(numeric_stats)
```

<h3>(i)

```python
correlation_matrix = numeric_features.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Coefficients Heatmap')
plt.show()
```
